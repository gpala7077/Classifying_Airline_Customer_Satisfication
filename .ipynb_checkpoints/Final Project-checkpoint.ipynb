# Initial Data Analysis

```python
import pandas as pd

# Combine test and train data
data = pd.concat([pd.read_csv('data/train.csv',index_col=0),pd.read_csv('data/test.csv',index_col=0)])
data.to_csv('data/aviation_data.csv')
data = data.set_index('id')

print('This dataset contains {} observations and {} attributes'.format(*data.shape))
```

```python
import matplotlib.pyplot as plt

# Show the distribution for all attributes
fig, axs = plt.subplots(4,6, figsize=(15,7))           # initialize 2 by 4 plot grid
fig.tight_layout()                                      # set layout
column_names = [column for column in data]              # Get column names into list
i = 0
for x in range(4):                                      # Fill plot grid
    for y in range(6):
        if i < data.shape[1]:
            axs[x,y].hist(data[column_names[i]])
            axs[x,y].set(title=column_names[i])
            i += 1
```
